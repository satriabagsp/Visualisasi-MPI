```python
# Read Data
df_data_kegiatan = func.read_clean_file('E:/OLAH DATA/PERENCANAAN/Eselon 1/Data/1. Es 1 2021.xlsx')

# Cleaning tanda baca
df_data_kegiatan['kegiatan_clean'] = df_data_kegiatan['kegiatan'].apply(func.cleaning_tanda_baca)

df_data_kegiatan
```

```python
# Read Data Indikator BPS
df_indikator = pd.read_csv('Data/Indikator BPS.csv', sep=';')

data = []
for index, row in df_indikator.iterrows():
    main_indicator = row['main_indicator']
    indicator = row['indicator']
    sub_indicator = row['sub_indicator']
    alias = row['alias']

    jumlah_subindikator = 0
    # Ganti tipe alias ke list
    alias = ast.literal_eval(alias)
    for a in alias:
        # Cari kata di setiap kegiatan
        jumlah = 0
        for kegiatan in df_data_kegiatan.kegiatan_clean:
            if a in kegiatan:
                jumlah += 1

        jumlah_subindikator += jumlah

        # print(f'{a} - {jumlah}')

    data_baris = [main_indicator, indicator, sub_indicator, jumlah_subindikator]

    # print(data_baris)

    data.append(data_baris)

# create dataframe
df_hasil = pd.DataFrame(data=data, columns=['Indikator Utama', 'Indikator', 'Sub-indikator', 'Jumlah Muncul'])

# Groupby indicator
df_hasil = df_hasil.groupby(by=['Indikator Utama', 'Indikator'], as_index=False).sum(['Jumlah Muncul'])

df_hasil
```
